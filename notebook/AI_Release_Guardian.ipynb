# ***AI Release Guardian — нейро-сотрудник для анализа веб-интерфейсов***

# Описание проекта

В данной работе реализован нейро-сотрудник **AI Release Guardian**, предназначенный для анализа веб-интерфейсов и выявления потенциальных проблем при выпуске новых версий продукта.

Основная задача системы — помочь QA-инженеру автоматически:

- анализировать структуру интерфейса веб-страницы  
- декомпозировать интерфейс на тестируемые элементы  
- группировать элементы по типам и классам эквивалентности  
- генерировать тестовые сценарии для элементов интерфейса  
- выявлять потенциальные зоны риска  
- анализировать изменения интерфейса между запусками  

Система выступает как **интеллектуальный помощник тестировщика**, который автоматически анализирует интерфейс страницы и формирует рекомендации по тестированию.

---

# Архитектурная идея

В отличие от классических RAG-чатботов, система использует **комбинированный подход анализа интерфейса**.

Источники информации:

1. **DOM-анализ веб-страницы**  
Система извлекает реальную структуру интерфейса страницы с помощью Selenium и BeautifulSoup.

2. **База знаний QA-практик**  
Используется RAG-архитектура для получения рекомендаций по тестированию.

3. **LLM-анализ интерфейса**  
LLM используется для reasoning-задач: анализа рисков, генерации тестов и интерпретации изменений.

Такой подход снижает вероятность галлюцинаций LLM, поскольку модель опирается на реальные данные интерфейса и структурированные артефакты анализа.

Архитектура построена по принципу:

- Code extracts facts → LLM performs reasoning

Это снижает вероятность галлюцинаций модели, поскольку LLM выполняет анализ на основе структурированных данных интерфейса.

# Основные компоненты системы

## 1. DOM-анализ

Система получает HTML страницы и извлекает элементы интерфейса.

Используются:

- Selenium — для загрузки страницы
- BeautifulSoup — для анализа DOM

Извлекаются основные элементы интерфейса:

- input  
- button  
- form  
- link  
- table  
- image  

---

## 2. Декомпозиция интерфейса

Интерфейс разбивается на **testable objects** — элементы, которые могут быть протестированы.

Пример:

```
input → поле ввода
button → кнопка
checkbox → переключатель
```
---

После этого элементы **группируются по типам**, и для каждого типа выбирается **representative element**, который используется для генерации тестов класса эквивалентных элементов.

---

## 3. RAG-слой

Для анализа используются знания из базы QA-практик.

Система:

1. извлекает релевантные документы  
2. передает их LLM  
3. использует полученный контекст при генерации тестов и анализе интерфейса  

Это позволяет использовать проверенные рекомендации по тестированию.

# Приемы улучшения RAG в проекте

В проекте используются следующие приемы улучшения RAG:

1. Grounded retrieval на основе реальных UI-элементов  
   Запрос к базе знаний формируется на основе DOM-атрибутов и testable objects.

2. Структурирование входных данных  
   Перед использованием LLM интерфейс переводится в testable objects, page_summary и diff_report, что уменьшает шум и снижает вероятность галлюцинаций.

3. Использование representative elements  
   Элементы группируются по типам и рассматриваются как классы эквивалентности, что уменьшает дублирование и повышает устойчивость генерации.

4. Многоуровневый анализ  
   Система использует element-level, page-level и release-level анализ вместо одного общего ответа.

5. Run-based архитектура и diff analysis  
   Контекст анализа расширяется за счёт истории запусков и сравнения версий интерфейса.

6. Гибридное использование LLM  
   Для демонстрации используется локальная русскоязычная модель, а для более сложного reasoning — внешний LLM API.

---

## 4. AI-анализ

LLM используется для задач reasoning:

- генерации тестовых сценариев  
- анализа UX-рисков  
- поиска потенциальных баг-зон  
- анализа изменений интерфейса между запусками  

В проекте используются два типа моделей:

- локальная русскоязычная LLM (экспериментальный слой анализа)  
- VSEGPT API (основной слой reasoning)

---

## 5. Run-based архитектура

Каждый запуск анализа сохраняется как **run**, что позволяет:

- хранить историю анализа интерфейса  
- сравнивать версии страницы  
- выявлять изменения между релизами  
- анализировать потенциальные зоны регрессии

## Форматы документов

Система поддерживает несколько форматов документов:

- PDF  
- DOCX

Это позволяет использовать разнообразные источники знаний при построении базы QA-практик.

````python
# ==========================================================
# Подготовка базы знаний
# ==========================================================
# В этом блоке:
# 1. Проверяем наличие папок базы знаний
# 2. Собираем пути ко всем документам
# 3. Выводим краткую статистику по базе знаний
# ==========================================================

import os
from glob import glob

# Пути к основным разделам базы знаний
theory_path = CONFIG["KB_THEORY_PATH"]
portfolio_path = CONFIG["KB_PORTFOLIO_PATH"]

print("Проверка папок базы знаний...\n")

for path in [theory_path, portfolio_path]:
    if os.path.exists(path):
        print(f"[OK] Папка найдена: {path}")
    else:
        print(f"[WARN] Папка не найдена: {path}")

# Поддерживаемые форматы документов
supported_extensions = ["*.pdf", "*.docx", "*.pptx", "*.txt"]

def collect_files_from_folder(folder_path, extensions):
    """
    Собирает список файлов из папки по заданным расширениям.

    Аргументы:
        folder_path (str): путь к папке
        extensions (list): список шаблонов расширений (*.pdf и т.д.)

    Возвращает:
        list: список найденных файлов
    """
    files = []
    for ext in extensions:
        files.extend(glob(os.path.join(folder_path, ext)))
    return sorted(files)

# Сбор документов из Theory и Portfolio
theory_files = collect_files_from_folder(theory_path, supported_extensions) if os.path.exists(theory_path) else []
portfolio_files = collect_files_from_folder(portfolio_path, supported_extensions) if os.path.exists(portfolio_path) else []

all_kb_files = theory_files + portfolio_files

print("\nСтатистика по базе знаний:")
print(f"Документов в Theory: {len(theory_files)}")
print(f"Документов в Portfolio: {len(portfolio_files)}")
print(f"Всего документов: {len(all_kb_files)}")

print("\nПримеры найденных файлов:")
for file_path in all_kb_files[:10]:
    print("-", file_path)
````

````python
# ==========================================================
# Проверка качества базы знаний
# ==========================================================
# Важно не только собрать документы, но и убедиться,
# что база знаний не пустая и содержит достаточный объём данных.
# ==========================================================

if len(all_kb_files) == 0:
    print("ВНИМАНИЕ: база знаний пуста. Проверьте содержимое папок Theory и Portfolio.")
elif len(all_kb_files) < 3:
    print("ПРЕДУПРЕЖДЕНИЕ: документов очень мало. Это может ухудшить качество retrieval.")
else:
    print("База знаний выглядит достаточной для демонстрации RAG-подхода.")
````

# Создание RAG-индекса

В проекте используется подход **Retrieval Augmented Generation (RAG)**.

Смысл этого подхода заключается в том, что перед генерацией рекомендаций система сначала ищет релевантные документы в базе знаний, а затем передаёт найденный контекст языковой модели.

В данном проекте RAG используется не как классический чат-бот, а как источник экспертного контекста для задач анализа интерфейсов.

LLM получает рекомендации по QA-практикам и использует их при:

- генерации тестовых сценариев
- анализе UX/UI рисков
- выявлении потенциальных баг-зон

---

## Почему используется RAG

Использование RAG позволяет:

- повысить точность рекомендаций
- снизить вероятность галлюцинаций
- привязать анализ к QA-домену
- использовать проверенные практики тестирования

Таким образом LLM выполняет reasoning на основе контекста из базы знаний.

---

## Этапы построения RAG-слоя

1. Загрузка документов из базы знаний  
2. Разбиение документов на фрагменты  
3. Построение эмбеддингов для каждого фрагмента  
4. Создание векторного индекса  
5. Настройка retriever для поиска релевантных фрагментов  

При запросе система извлекает наиболее релевантные фрагменты и передаёт их LLM в качестве контекста.

---

## Используемые технологии

- **LlamaIndex** — построение RAG-индекса  
- **HuggingFace Embeddings** — генерация эмбеддингов  
- **sentence-transformers** — базовая модель эмбеддингов

````python
# ==========================================================
# Создание RAG-индекса
# ==========================================================
# В этом блоке:
# 1. Загружаем документы базы знаний
# 2. Настраиваем модель эмбеддингов
# 3. Строим векторный индекс
# 4. Создаём retriever для дальнейшего поиска
# ==========================================================

from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

# Настройка модели эмбеддингов
print("Загрузка модели эмбеддингов...")
embed_model = HuggingFaceEmbedding(model_name=CONFIG["EMBEDDING_MODEL"])

# Передаём модель эмбеддингов в глобальные настройки LlamaIndex
Settings.embed_model = embed_model

print("Модель эмбеддингов успешно настроена:")
print(CONFIG["EMBEDDING_MODEL"])
````

````python
# ==========================================================
# Загрузка документов базы знаний
# ==========================================================
# LlamaIndex может читать документы прямо из папок.
# Мы отдельно читаем Theory и Portfolio, затем объединяем их.
# ==========================================================

documents = []

# Загружаем Theory
if os.path.exists(theory_path):
    print("Загрузка документов из Theory...")
    theory_docs = SimpleDirectoryReader(theory_path).load_data()
    print(f"Загружено документов из Theory: {len(theory_docs)}")
    documents.extend(theory_docs)
else:
    print("Папка Theory не найдена")

# Загружаем Portfolio
if os.path.exists(portfolio_path):
    print("Загрузка документов из Portfolio...")
    portfolio_docs = SimpleDirectoryReader(portfolio_path).load_data()
    print(f"Загружено документов из Portfolio: {len(portfolio_docs)}")
    documents.extend(portfolio_docs)
else:
    print("Папка Portfolio не найдена")

print(f"\nВсего загружено документов: {len(documents)}")
````

````python
# ==========================================================
# Ограничение числа документов для ускорения Colab
# ==========================================================
# В учебной среде Colab слишком большой объём документов
# может замедлить построение индекса.
# Поэтому используем ограничение из CONFIG.
# ==========================================================

if len(documents) > CONFIG["MAX_DOCUMENTS"]:
    documents = documents[:CONFIG["MAX_DOCUMENTS"]]
    print(f"Количество документов ограничено до {CONFIG['MAX_DOCUMENTS']} для ускорения работы.")

print(f"Документов будет использовано для индексации: {len(documents)}")
````

````python
# ==========================================================
# Построение векторного индекса
# ==========================================================
# Индекс позволяет искать релевантные документы
# по смыслу, а не только по ключевым словам.
# ==========================================================

print("Начинаем построение VectorStoreIndex...")

index = VectorStoreIndex.from_documents(documents)

print("VectorStoreIndex успешно создан.")
````

````python
# ==========================================================
# Создание retriever
# ==========================================================
# Retriever — это объект, который будет искать
# наиболее релевантные фрагменты документов.
# ==========================================================

retriever = index.as_retriever(
    similarity_top_k=CONFIG["TOP_K_RETRIEVAL"]
)

print("Retriever успешно создан.")
print(f"Параметр similarity_top_k = {CONFIG['TOP_K_RETRIEVAL']}")
````

# **Фильтрация retrieval-контекста**

После извлечения фрагментов из базы знаний система выполняет дополнительную фильтрацию контекста.
Это позволяет уменьшить шум retrieval, убрать слишком общие фрагменты и передавать LLM только наиболее полезную информацию.

````python
# ==========================================================
# CONTEXT FILTERING FOR RAG
# ==========================================================

def filter_retrieved_context(
    nodes,
    max_total_chars=1500,
    min_text_len=80,
    max_chunk_len=1400
):
    """
    Фильтрует retrieved nodes перед передачей в LLM.

    Логика:
    - убирает слишком короткие фрагменты
    - пытается убрать слишком общие QA-фрагменты
    - ограничивает общий размер контекста
    - если всё отфильтровалось, возвращает fallback-контекст
    """

    filtered_chunks = []
    total_chars = 0

    GENERIC_PATTERNS = [
        "кто такой тестировщик",
        "что такое тестирование",
        "миф о профессии",
        "защитник качества",
        "test plan identifier",
        "features to be tested",
        "features not to be tested",
        "references",
        "approach",
        "testlink",
        "execution type",
        "estimated exec. duration"
    ]

    for node in nodes:
        node_text = getattr(node, "text", None)

        if node_text is None and hasattr(node, "node"):
            node_text = getattr(node.node, "text", "")

        node_text = str(node_text).strip()

        if len(node_text) < min_text_len:
            continue

        lower_text = node_text.lower()

        # Если chunk слишком общий, но не весь состоит из мусора,
        # не выбрасываем его сразу полностью — просто пропускаем
        # только самые явно общие документы
        generic_hits = sum(pattern in lower_text for pattern in GENERIC_PATTERNS)

        if generic_hits >= 3:
            continue

        # Слишком длинные chunks режем, а не выбрасываем
        if len(node_text) > max_chunk_len:
            node_text = node_text[:max_chunk_len]

        if total_chars + len(node_text) > max_total_chars:
            remaining = max_total_chars - total_chars
            if remaining > min_text_len:
                filtered_chunks.append(node_text[:remaining])
            break

        filtered_chunks.append(node_text)
        total_chars += len(node_text)

    # ------------------------------------------------------
    # fallback: если всё отфильтровалось, берём первые
    # укороченные chunks без жёсткой фильтрации
    # ------------------------------------------------------
    if not filtered_chunks:
        fallback_chunks = []
        total_chars = 0

        for node in nodes:
            node_text = getattr(node, "text", None)

            if node_text is None and hasattr(node, "node"):
                node_text = getattr(node.node, "text", "")

            node_text = str(node_text).strip()

            if len(node_text) < min_text_len:
                continue

            node_text = node_text[:500]

            if total_chars + len(node_text) > max_total_chars:
                break

            fallback_chunks.append(node_text)
            total_chars += len(node_text)

        print("⚠️ Жёсткая фильтрация убрала весь контекст. Используется fallback-контекст.")
        return "\n\n".join(fallback_chunks)

    return "\n\n".join(filtered_chunks)
````

# Улучшение RAG: расширение запроса (Query Enrichment)

Чтобы повысить точность retrieval, запрос к базе знаний расширяется
дополнительными QA-терминами.

Это позволяет retriever находить более релевантные рекомендации
по тестированию интерфейсных элементов.

````python
# ==========================================================
# RAG QUERY ENRICHMENT
# ==========================================================

QA_TERMS = [
    "валидация поля",
    "негативные сценарии",
    "граничные значения",
    "пограничные случаи",
    "проверка ошибок",
    "поведение интерфейса",
    "валидация формы",
    "проверка ввода пользователя"
]

def enrich_query(base_query, element_type=None):
    """
    Расширяет запрос к базе знаний для улучшения retrieval.
    Делает запрос более узким и ориентированным на тестирование UI-элемента.
    """

    terms = ", ".join(QA_TERMS)

    if element_type:
        enriched = (
            f"{base_query}\n\n"
            f"Сфокусируйся на рекомендациях по тестированию элемента интерфейса типа: {element_type}. "
            f"Приоритет отдавай проверкам валидации поля, негативным сценариям, "
            f"граничным значениям, поведению интерфейса и обработке ошибок. "
            f"Игнорируй общие тест-планы и бизнес-сценарии вроде корзины, заказов, "
            f"подарочных карт или профилей пользователей. "
            f"Используй подходы QA: {terms}."
        )
    else:
        enriched = (
            f"{base_query}\n\n"
            f"Сфокусируйся на рекомендациях по тестированию элементов интерфейса. "
            f"Предпочитай советы по валидации ввода, негативным сценариям, "
            f"граничным значениям и проверкам UI-поведения. "
            f"Игнорируй общие тест-планы бизнес-систем. "
            f"Используй QA-подходы: {terms}."
        )

    return enriched
````

# Улучшение RAG: фильтрация результатов retrieval

Некоторые найденные фрагменты могут быть слабо релевантными.

Поэтому используется дополнительная фильтрация результатов
по score similarity.

````python
# ==========================================================
# SCORE FILTERING
# ==========================================================

SIMILARITY_THRESHOLD = 0.1

def filter_by_score(nodes):
    """
    Отсеивает слишком слабые retrieval results по score.
    Если после фильтрации список пуст, возвращает исходные nodes.
    """

    filtered = []

    for node in nodes:
        score = getattr(node, "score", None)

        if score is None:
            filtered.append(node)
            continue

        if score >= SIMILARITY_THRESHOLD:
            filtered.append(node)

    # Если всё отфильтровалось, возвращаем исходный список
    if not filtered:
        print("Все узлы были отфильтрованы по score. Возвращаем исходные retrieval results.")
        return nodes

    return filtered
````

# Улучшение RAG: простой reranking контекста

После retrieval применяется дополнительная сортировка контекста
по наличию QA-ключевых слов.

Это помогает поднимать более полезные рекомендации выше
в передаваемом контексте.

````python
# ==========================================================
# LIGHT RERANKING
# ==========================================================

QA_KEYWORDS = [
    "валидация поля",
    "негативные сценарии",
    "граничные значения",
    "пограничные случаи",
    "проверка ошибок",
    "поведение интерфейса",
    "валидация формы",
    "проверка ввода пользователя"
]

def rerank_nodes(nodes):

    def score_node(node):
        text = getattr(node, "text", "").lower()

        score = 0
        for kw in QA_KEYWORDS:
            if kw in text:
                score += 1

        return score

    ranked = sorted(nodes, key=score_node, reverse=True)

    return ranked
````

````python
# ==========================================================
# RAG TRACE DEMO — шаг 3
# ==========================================================

enriched_query_text = enrich_query(rag_query)

retrieved_nodes = retriever.retrieve(enriched_query_text)

retrieved_nodes = filter_by_score(retrieved_nodes)

retrieved_nodes = rerank_nodes(retrieved_nodes)

print(f"Количество найденных фрагментов: {len(retrieved_nodes)}")
````

````python
# ==========================================================
# RAG TRACE DEMO — шаг 4
# Просмотр найденного контекста
# ==========================================================
# Здесь мы выводим несколько найденных фрагментов,
# чтобы увидеть, какой контекст система будет передавать LLM.
# ==========================================================

print("\nSTEP 4 — RETRIEVED CONTEXT\n")

for i, node in enumerate(retrieved_nodes[:CONFIG["TOP_K_RETRIEVAL"]], start=1):
    print("=" * 80)
    print(f"Фрагмент контекста #{i}")
    print("=" * 80)

    # У разных версий узлов текст может лежать в разных полях.
    # Поэтому используем безопасное извлечение текста.
    node_text = getattr(node, "text", None)
    if node_text is None and hasattr(node, "node"):
        node_text = getattr(node.node, "text", "")

    print(str(node_text)[:1500])
    print("\n")
````

````python
# ==========================================================
# RAG TRACE DEMO — шаг 5
# Подготовка контекста для LLM-анализа
# ==========================================================
# На этом этапе найденные фрагменты проходят дополнительную
# фильтрацию, чтобы уменьшить шум retrieval и передать LLM
# только наиболее полезный контекст.
# ==========================================================

print("\nSTEP 5 — CONTEXT ASSEMBLY\n")

retrieved_context = filter_retrieved_context(retrieved_nodes)

print("Собранный контекст для LLM после фильтрации:")
print(retrieved_context[:3000])
````

# Run-based архитектура анализа

AI Release Guardian сохраняет результаты анализа сайта в виде отдельных **runs**.

Каждый запуск анализа создаёт новую папку, содержащую артефакты анализа интерфейса.

В run сохраняются:

- HTML страницы  
- testable objects  
- сгенерированные тестовые сценарии  
- результаты AI-анализа  
- отчёты об изменениях интерфейса  
- итоговая сводка анализа  

Такая структура позволяет сохранять состояние анализа на момент запуска.

---

## Преимущества run-based архитектуры

Использование отдельных run-папок позволяет:

- хранить историю анализа интерфейса  
- сравнивать версии страницы между запусками  
- выявлять изменения интерфейса (UI diff)  
- анализировать потенциальные регрессии  

Таким образом система может использовать предыдущие результаты анализа для оценки рисков при выпуске новых версий интерфейса.

````python
# ==========================================================
# RUN ARCHITECTURE
# ==========================================================
# Каждый запуск анализа сохраняется в отдельную папку
# внутри Google Drive.
# Это позволяет хранить историю анализов сайта
# между сессиями Colab.
# ==========================================================

import os
import re
from urllib.parse import urlparse

def make_safe_site_name(url: str) -> str:
    """
    Преобразует URL сайта в безопасное имя папки.
    """
    parsed = urlparse(url)
    raw_name = parsed.netloc or "unknown_site"
    safe_name = raw_name.replace(".", "_").replace(":", "_")
    return safe_name

# TARGET_URL уже выбран на предыдущем шаге
site_name = make_safe_site_name(TARGET_URL)

print("Сайт:", site_name)

# Корневая папка для run-артефактов в Google Drive
sites_root = CONFIG["RUNS_ROOT"]

# Папка конкретного сайта
site_dir = os.path.join(sites_root, site_name)

# Папка всех запусков
runs_dir = os.path.join(site_dir, "runs")

# Создаём директории
os.makedirs(runs_dir, exist_ok=True)

# Ищем уже существующие run
existing_runs = [
    d for d in os.listdir(runs_dir)
    if re.match(r"run_\d{3}$", d)
]

# Определяем следующий номер run
if len(existing_runs) == 0:
    run_number = 1
else:
    run_numbers = [int(r.split("_")[1]) for r in existing_runs]
    run_number = max(run_numbers) + 1

current_run_name = f"run_{run_number:03d}"
run_dir = os.path.join(runs_dir, current_run_name)

os.makedirs(run_dir, exist_ok=True)

# Все артефакты текущего запуска будут сохраняться сюда
artifacts_dir = run_dir

print("Создан новый run:")
print(run_dir)
````

# Декомпозиция интерфейса на testable objects

После получения HTML страницы необходимо перейти от сырого DOM к более удобному представлению интерфейса.

Для этого в проекте используется слой **testable objects** — нормализованных объектов интерфейса, пригодных для тестирования и дальнейшего AI-анализа.

---

## Зачем нужна декомпозиция

Сырой HTML неудобно напрямую передавать в LLM, потому что:

- в нём содержится большое количество служебной информации  
- элементы представлены в низкоуровневом виде  
- анализ становится менее устойчивым  

Поэтому система выполняет **нормализацию DOM** и переводит элементы страницы в структурированные тестируемые сущности.

---

## Примеры testable objects

К таким объектам относятся:

- поле ввода  
- кнопка  
- чекбокс  
- радиокнопка  
- таблица  
- изображение  
- ссылка  
- другие интерактивные элементы интерфейса  

---

## Классы эквивалентности элементов

После извлечения элементов система группирует их по типам интерфейсных объектов.

Например:

- все кнопки  
- все поля ввода  
- все ссылки  

Элементы одного типа обычно тестируются **по одинаковым принципам**, поэтому они рассматриваются как **класс эквивалентности**.

Для генерации тестов выбирается **representative element** — один типичный представитель группы.

Это позволяет:

- сократить объём генерации тестов  
- избежать дублирования  
- сохранить покрытие основных сценариев тестирования.

---

## Что происходит на этом этапе

На этапе декомпозиции выполняются следующие действия:

- выделение основных типов элементов из DOM  
- извлечение полезных атрибутов элементов  
- формирование нормализованного списка testable objects  
- группировка элементов по типам  
- подготовка summary по странице

````python
# ==========================================================
# Декомпозиция DOM → testable objects
# ==========================================================
# В этом блоке:
# 1. Используем уже полученный HTML страницы
# 2. Парсим его через BeautifulSoup
# 3. Извлекаем тестируемые элементы интерфейса
# 4. Приводим их к единой нормализованной структуре
# ==========================================================

from bs4 import BeautifulSoup
import json
import os

# Повторно парсим HTML, если soup ещё не создан
soup = BeautifulSoup(page_html, "html.parser")

print("DOM готов к декомпозиции")
````

````python
# ==========================================================
# Вспомогательные функции для извлечения атрибутов
# ==========================================================

def safe_get_attr(tag, attr_name, default=""):
    """
    Безопасно получает значение атрибута HTML-тега.

    Аргументы:
        tag: объект BeautifulSoup tag
        attr_name (str): имя атрибута
        default (str): значение по умолчанию

    Возвращает:
        str: значение атрибута или default
    """
    if tag is None:
        return default
    return tag.attrs.get(attr_name, default)


def normalize_text(value):
    """
    Нормализует текст:
    - убирает лишние пробелы
    - приводит None к пустой строке

    Аргументы:
        value: входное значение

    Возвращает:
        str: очищенный текст
    """
    if value is None:
        return ""
    return " ".join(str(value).split()).strip()


def build_display_name(obj):
    """
    Формирует человекочитаемое имя объекта.
    Используется для отображения в интерфейсе и отчётах.
    """
    candidates = [
        obj.get("label", ""),
        obj.get("name", ""),
        obj.get("id", ""),
        obj.get("placeholder", ""),
        obj.get("text", ""),
        obj.get("locator", "")
    ]

    for item in candidates:
        item = normalize_text(item)
        if item:
            return item

    return f"{obj.get('object_type', 'unknown_object')}_{obj.get('index', 0)}"
````

````python
# ==========================================================
# Контейнер для testable objects
# ==========================================================

testable_objects = []

def add_object(object_type, tag, index, extra=None):
    """
    Добавляет нормализованный объект интерфейса в общий список.

    Аргументы:
        object_type (str): тип объекта (input, button и т.д.)
        tag: HTML-тег BeautifulSoup
        index (int): порядковый индекс объекта
        extra (dict): дополнительные поля
    """
    obj = {
        "index": index,
        "object_type": object_type,
        "tag_name": tag.name if tag else "",
        "id": normalize_text(safe_get_attr(tag, "id")),
        "name": normalize_text(safe_get_attr(tag, "name")),
        "type": normalize_text(safe_get_attr(tag, "type")),
        "class": " ".join(tag.get("class", [])) if tag and tag.get("class") else "",
        "placeholder": normalize_text(safe_get_attr(tag, "placeholder")),
        "value": normalize_text(safe_get_attr(tag, "value")),
        "text": normalize_text(tag.get_text(" ", strip=True) if tag else ""),
        "aria_label": normalize_text(safe_get_attr(tag, "aria-label")),
        "title": normalize_text(safe_get_attr(tag, "title")),
        "href": normalize_text(safe_get_attr(tag, "href")),
        "src": normalize_text(safe_get_attr(tag, "src")),
        "locator": ""
    }

    # Простейший CSS-подобный locator для удобства анализа
    if obj["id"]:
        obj["locator"] = f"#{obj['id']}"
    elif obj["name"]:
        obj["locator"] = f'[name="{obj["name"]}"]'
    else:
        obj["locator"] = tag.name if tag else "unknown"

    # Дополнительные поля, если они есть
    if extra:
        obj.update(extra)

    # Формируем display_name после заполнения полей
    obj["display_name"] = build_display_name(obj)

    testable_objects.append(obj)
````

````python
# ==========================================================
# Извлечение input-элементов
# ==========================================================
# Сюда входят:
# - text input
# - password input
# - email input
# - hidden, checkbox, radio и т.д.
# В дальнейшем checkbox и radio можно анализировать отдельно,
# но на этом шаге мы сначала собираем все input.
# ==========================================================

input_tags = soup.find_all("input")

for idx, tag in enumerate(input_tags, start=1):
    add_object("input", tag, idx)

print(f"Найдено input-элементов: {len(input_tags)}")
````

````python
# ==========================================================
# Извлечение button-элементов
# ==========================================================

button_tags = soup.find_all("button")

for idx, tag in enumerate(button_tags, start=1):
    add_object("button", tag, idx)

print(f"Найдено button-элементов: {len(button_tags)}")
````

````python
# ==========================================================
# Извлечение form-элементов
# ==========================================================

form_tags = soup.find_all("form")

for idx, tag in enumerate(form_tags, start=1):
    add_object("form", tag, idx)

print(f"Найдено form-элементов: {len(form_tags)}")
````

````python
# ==========================================================
# Извлечение ссылок
# ==========================================================

link_tags = soup.find_all("a")

for idx, tag in enumerate(link_tags, start=1):
    add_object("link", tag, idx)

print(f"Найдено ссылок: {len(link_tags)}")
````

````python
# ==========================================================
# Извлечение изображений
# ==========================================================

image_tags = soup.find_all("img")

for idx, tag in enumerate(image_tags, start=1):
    add_object("image", tag, idx)

print(f"Найдено изображений: {len(image_tags)}")
````

````python
# ==========================================================
# Извлечение таблиц
# ==========================================================

table_tags = soup.find_all("table")

for idx, tag in enumerate(table_tags, start=1):
    add_object("table", tag, idx)

print(f"Найдено таблиц: {len(table_tags)}")
````

````python
# ==========================================================
# Дополнительная классификация input-элементов
# ==========================================================
# После первичного сбора input мы можем уточнить их тип.
# Это полезно для QA-анализа, потому что password, checkbox
# и radio требуют разных тестовых сценариев.
# ==========================================================

for obj in testable_objects:
    if obj["object_type"] == "input":
        input_type = obj.get("type", "").lower()

        if input_type == "checkbox":
            obj["object_type"] = "checkbox"
        elif input_type == "radio":
            obj["object_type"] = "radio"
        elif input_type == "password":
            obj["object_type"] = "password_input"
        elif input_type == "email":
            obj["object_type"] = "email_input"
        elif input_type == "hidden":
            obj["object_type"] = "hidden_input"
````

````python
# ==========================================================
# Сводка по типам testable objects
# ==========================================================

from collections import Counter

object_type_counter = Counter(obj["object_type"] for obj in testable_objects)

print("Сводка по testable objects:\n")
for object_type, count in object_type_counter.items():
    print(f"{object_type}: {count}")

print(f"\nВсего testable objects: {len(testable_objects)}")
````

````python
# ==========================================================
# Просмотр первых testable objects
# ==========================================================
# Это помогает визуально проверить, что декомпозиция
# выполнена корректно.
# ==========================================================

print("Первые testable objects:\n")

for obj in testable_objects[:10]:
    print(json.dumps(obj, ensure_ascii=False, indent=2))
    print("-" * 80)
````

````python
# ==========================================================
# Сохранение testable objects в JSON
# ==========================================================
# Сохраняем результат как отдельный артефакт анализа.
# ==========================================================

artifacts_dir = run_dir
os.makedirs(artifacts_dir, exist_ok=True)

testable_objects_path = os.path.join(artifacts_dir, "testable_objects.json")

with open(testable_objects_path, "w", encoding="utf-8") as f:
    json.dump(testable_objects, f, ensure_ascii=False, indent=2)

print("Testable objects сохранены:")
print(testable_objects_path)
````

# Выбор representative elements по типам объектов

После декомпозиции интерфейса система получает список **testable objects**.

Многие элементы интерфейса относятся к одному и тому же типу (например, несколько кнопок или несколько полей ввода).

Чтобы уменьшить дублирование тестов, элементы рассматриваются как **классы эквивалентности**.

Элементы одного типа обычно тестируются по одинаковым принципам, поэтому нет необходимости генерировать отдельный набор тестов для каждого элемента.

---

## Representative element

На этом этапе для каждого типа объектов выбирается один **representative element** — типичный представитель группы.

Этот элемент используется для генерации тестовых сценариев для всего класса элементов.

Например:


````python
# ==========================================================
# Representative elements по типам объектов
# ==========================================================
# Эта ячейка:
# 1. Берёт уже существующий список testable_objects
# 2. Группирует элементы по object_type
# 3. Выбирает по одному representative element на каждый тип
#
# ВАЖНО:
# - старая логика selected_object НЕ меняется
# - testable_objects НЕ перезаписывается
# - мы только добавляем новый слой данных
# ==========================================================

from collections import defaultdict
import json

# Проверяем, что testable_objects уже существует
if "testable_objects" not in globals() or not testable_objects:
    raise Exception(
        "Список testable_objects не найден или пуст. "
        "Сначала выполни декомпозицию DOM."
    )

# Группировка объектов по типу
objects_by_type = defaultdict(list)

for obj in testable_objects:
    obj_type = obj.get("object_type", "unknown")
    objects_by_type[obj_type].append(obj)

# Выбор representative element:
# берём первый объект каждого типа
representative_objects = {}

for obj_type, objects in objects_by_type.items():
    representative_objects[obj_type] = objects[0]

print("Representative elements по типам объектов:\n")

for obj_type, obj in representative_objects.items():
    print(
        f"- {obj_type}: "
        f"{obj.get('display_name', '')} | "
        f"{obj.get('locator', '')}"
    )

print(f"\nВсего найдено типов объектов: {len(representative_objects)}")
````

````python
# ==========================================================
# Сохранение representative elements по типам объектов
# ==========================================================
# Эта ячейка:
# 1. Сохраняет representative_objects в текущую run-папку
# 2. Не меняет старые артефакты
# 3. Готовит базу для генерации тестов по типам
# ==========================================================

import os
import json

# Проверяем, что representative_objects уже существует
if "representative_objects" not in globals() or not representative_objects:
    raise Exception(
        "representative_objects не найден. "
        "Сначала выполни ячейку выбора representative elements."
    )

representative_objects_path = os.path.join(
    artifacts_dir,
    "representative_objects_by_type.json"
)

with open(representative_objects_path, "w", encoding="utf-8") as f:
    json.dump(representative_objects, f, ensure_ascii=False, indent=2)

print("Representative elements сохранены:")
print(representative_objects_path)
````

# Генерация тестов для representative elements

Для каждого типа интерфейсных объектов используется один **representative element**.  
На его основе система генерирует тестовые сценарии для всего класса эквивалентных элементов.

Тесты формируются на основе:

- структуры элемента интерфейса  
- рекомендаций из базы знаний QA-практик (RAG)  
- reasoning языковой модели  

Таким образом LLM выполняет анализ типичного элемента и предлагает тестовые сценарии, применимые ко всем элементам данного типа.

## Преимущества такого подхода

````python
# ==========================================================
# Helper функция для вызова VSEGPT API
# ==========================================================

import requests

# ==========================================================
# Helper функция для вызова VSEGPT API
# ==========================================================

import requests

def call_vsegpt_api(system_prompt, user_prompt, model="openai/gpt-4o-mini"):

    url = "https://api.vsegpt.ru/v1/chat/completions"

    headers = {
        "Authorization": f"Bearer {VSEGPT_API_KEY}",
        "Content-Type": "application/json"
    }

    safe_system_prompt = safety_filter(system_prompt)
    safe_user_prompt = safety_filter(user_prompt)

    payload = {
        "model": model,
        "messages": [
            {"role": "system", "content": safe_system_prompt},
            {"role": "user", "content": safe_user_prompt}
        ],
        "temperature": 0.3
    }

    response = requests.post(url, headers=headers, json=payload)

    if response.status_code != 200:
        raise Exception(f"Ошибка VSEGPT API: {response.text}")

    data = response.json()

    return data["choices"][0]["message"]["content"]
````

**Безопасный слой обработки запросов**

Перед передачей пользовательских или автоматически сформированных запросов в LLM система выполняет базовую фильтрацию.
Это позволяет снизить риск prompt injection и некорректной интерпретации входных данных.

````python
# ==========================================================
# SAFETY FILTER
# ==========================================================

ENABLE_SAFETY_FILTER = True

BLOCKED_PATTERNS = [
    "ignore previous instructions",
    "ignore all previous instructions",
    "system prompt",
    "reveal system prompt",
    "show hidden instructions",
    "delete all data",
    "execute code",
    "run shell",
    "os.system",
    "subprocess",
    "rm -rf",
]

def safety_filter(user_input: str) -> str:
    """
    Фильтрация prompt перед передачей в LLM.
    """

    if not ENABLE_SAFETY_FILTER:
        return user_input

    text = str(user_input).lower()

    for pattern in BLOCKED_PATTERNS:

        if pattern in text:

            print("⚠️ SAFETY FILTER TRIGGERED")
            print("Blocked pattern:", pattern)

            raise ValueError(
                f"Запрос заблокирован системой безопасности: найден '{pattern}'"
            )

    return user_input
````

````python
# ==========================================================
# Helper-функция для вызова VSEGPT API
# ==========================================================
# Эта функция будет использоваться в тех блоках,
# где локальная модель слишком медленная или нестабильная.
# ==========================================================

import requests
import json

def call_vsegpt_api(system_prompt: str, user_prompt: str, model: str = "openai/gpt-4o-mini"):
    """
    Выполняет запрос к VSEGPT API.

    Аргументы:
        system_prompt (str): системный промпт
        user_prompt (str): пользовательский промпт
        model (str): имя модели

    Возвращает:
        str: текст ответа модели
    """
    url = "https://api.vsegpt.ru/v1/chat/completions"

    headers = {
        "Authorization": f"Bearer {VSEGPT_API_KEY}",
        "Content-Type": "application/json"
    }

    payload = {
        "model": model,
        "messages": [
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        "temperature": 0.3
    }

    response = requests.post(url, headers=headers, json=payload, timeout=120)
    response.raise_for_status()

    data = response.json()
    return data["choices"][0]["message"]["content"]
````

 **Helper для локальной русскоязычной LLM**

Для повышения качества генерации локальная модель вызывается через единый helper, использующий chat template и строгую структуру ответа.

````python
# ==========================================================
# Helper для локальной русскоязычной LLM
# ==========================================================

import torch

def call_local_ru_llm(system_prompt: str, user_prompt: str, max_new_tokens: int = 250):
    """
    Унифицированный вызов локальной русскоязычной LLM.
    Использует chat template и возвращает только новый сгенерированный текст.
    """

    safe_system_prompt = safety_filter(system_prompt)
    safe_user_prompt = safety_filter(user_prompt)

    messages = [
        {
            "role": "system",
            "content": safe_system_prompt
        },
        {
            "role": "user",
            "content": safe_user_prompt
        }
    ]

    text = ru_tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    inputs = ru_tokenizer(text, return_tensors="pt")

    if torch.cuda.is_available():
        inputs = {k: v.to("cuda") for k, v in inputs.items()}
    else:
        inputs = {k: v.to(ru_model.device) for k, v in inputs.items()}

    outputs = ru_model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        do_sample=False,
        pad_token_id=ru_tokenizer.eos_token_id,
        eos_token_id=ru_tokenizer.eos_token_id
    )

    generated_tokens = outputs[0][inputs["input_ids"].shape[1]:]
    answer = ru_tokenizer.decode(generated_tokens, skip_special_tokens=True).strip()

    return answer
````

````python
# ==========================================================
# AI-анализ рисков страницы через локальную русскоязычную LLM
# ==========================================================
# Для локальной модели используем короткий и строго структурированный prompt.
# ==========================================================

page_risk_prompt_llm = f"""
Сводка страницы:
{json.dumps(page_summary, ensure_ascii=False, indent=2)}

Сделай анализ рисков страницы.

Ответь строго по шаблону:

1. Основные риски:
2. Возможные зоны регрессии:
3. Что проверить в первую очередь:
4. Возможные UX-проблемы:
5. Достаточность данных:

Пиши кратко и по существу.
Не повторяй инструкцию.
Не пиши фразы вроде "делай выводы на основе данных".
"""

page_risk_analysis_llm = call_local_ru_llm(
    system_prompt=(
        "Ты опытный QA Lead. "
        "Анализируй интерфейс страницы по предоставленной summary. "
        "Отвечай строго по шаблону, кратко, на русском языке."
    ),
    user_prompt=page_risk_prompt_llm,
    max_new_tokens=420
)

print("Результат AI-анализа рисков страницы:\n")
print(page_risk_analysis_llm)
````

# Diff-анализ интерфейса между запусками

Одной из задач **AI Release Guardian** является обнаружение изменений интерфейса между версиями сайта.

Каждый запуск анализа сохраняет результаты DOM-декомпозиции в виде файла **testable_objects.json**.

Сравнивая такие файлы между разными run, система может определить:

- какие элементы появились  
- какие элементы исчезли  
- какие элементы изменились  

Это позволяет выявлять потенциальные зоны регрессии и планировать дополнительное тестирование.

---

## Роль diff-анализа

Diff-анализ позволяет перейти от анализа отдельной страницы к **анализу изменений интерфейса между версиями**.

Таким образом система может обнаруживать изменения UI, которые могут повлиять на пользовательские сценарии.

---

## Release-level анализ

Diff между run используется как основа для **release-level reasoning**, где система оценивает:

- потенциальные регрессии  
- новые риски интерфейса  
- области, требующие дополнительного тестирования

Это делает AI Release Guardian инструментом поддержки QA-процесса при выпуске новых версий интерфейса.

````python
# ==========================================================
# Диагностика путей run
# ==========================================================

import os

print("site_name:", site_name)
print("runs_dir:", runs_dir)
print("run_dir:", run_dir)
print("current_run_name:", current_run_name)

print("\nПапки в runs_dir:")
if os.path.exists(runs_dir):
    for item in sorted(os.listdir(runs_dir)):
        print("-", item)
else:
    print("runs_dir не существует")
````

````python
# ==========================================================
# Выбор предыдущего и текущего run для diff-анализа
# ==========================================================

import os
import re

# Гарантируем, что runs_dir соответствует run_dir
runs_dir = os.path.dirname(run_dir)

all_runs = sorted([
    d for d in os.listdir(runs_dir)
    if re.match(r"run_\d{3}$", d)
])

print("Найденные run:")
print(all_runs)

if current_run_name not in all_runs:
    raise Exception(
        f"Текущий run {current_run_name} не найден в runs_dir: {runs_dir}"
    )

current_index = all_runs.index(current_run_name)

if current_index == 0:
    raise Exception("Для diff-анализа нужен хотя бы один предыдущий run.")

run_old = all_runs[current_index - 1]
run_new = current_run_name

print("Сравниваем:")
print("OLD:", run_old)
print("NEW:", run_new)
````

````python
# ==========================================================
# Загрузка testable_objects двух запусков
# ==========================================================

import json
import os

# runs_dir уже создан в блоке RUN ARCHITECTURE
# и указывает на папку runs конкретного сайта

path_old = os.path.join(runs_dir, run_old, "testable_objects.json")
path_new = os.path.join(runs_dir, run_new, "testable_objects.json")

print("OLD path:", path_old)
print("NEW path:", path_new)

# Загружаем данные
with open(path_old, "r", encoding="utf-8") as f:
    objects_old = json.load(f)

with open(path_new, "r", encoding="utf-8") as f:
    objects_new = json.load(f)

print("OLD objects:", len(objects_old))
print("NEW objects:", len(objects_new))
````

````python
# ==========================================================
# Индексирование элементов по locator
# ==========================================================

old_index = {obj["locator"]: obj for obj in objects_old}
new_index = {obj["locator"]: obj for obj in objects_new}

old_locators = set(old_index.keys())
new_locators = set(new_index.keys())
````

````python
# ==========================================================
# Diff-анализ
# ==========================================================

added_elements = new_locators - old_locators
removed_elements = old_locators - new_locators
common_elements = old_locators & new_locators

changed_elements = []

for locator in common_elements:

    old_obj = old_index[locator]
    new_obj = new_index[locator]

    if old_obj != new_obj:
        changed_elements.append(locator)

print("Добавленные элементы:", len(added_elements))
print("Удалённые элементы:", len(removed_elements))
print("Изменённые элементы:", len(changed_elements))
````

````python
print("\nДобавленные элементы:")
for loc in list(added_elements)[:10]:
    print(loc)

print("\nУдалённые элементы:")
for loc in list(removed_elements)[:10]:
    print(loc)

print("\nИзменённые элементы:")
for loc in changed_elements[:10]:
    print(loc)
````

````python
# ==========================================================
# Формирование diff-отчёта
# ==========================================================

diff_report = {
    "site": site_name,
    "old_run": run_old,
    "new_run": run_new,
    "added_elements": list(added_elements),
    "removed_elements": list(removed_elements),
    "changed_elements": changed_elements
}

print(json.dumps(diff_report, ensure_ascii=False, indent=2))
````

````python
# ==========================================================
# Сохранение diff-отчёта
# ==========================================================

diff_path = os.path.join(
    runs_dir,
    run_new,
    "ui_diff_report.json"
)

with open(diff_path, "w", encoding="utf-8") as f:
    json.dump(diff_report, f, ensure_ascii=False, indent=2)

print("Diff-отчёт сохранён:")
print(diff_path)
````

# Оценка риска релиза

На основе diff интерфейса между двумя запусками система вычисляет простую оценку риска релиза (**risk score**).

Оценка строится на основе изменений структуры интерфейса между версиями страницы.

Risk score зависит от количества:

- добавленных элементов  
- удалённых элементов  
- изменённых элементов  

---

## Интерпретация risk score

Чем больше изменений обнаружено в интерфейсе, тем выше потенциальный риск регрессии.

Такая метрика позволяет быстро оценить, насколько изменения интерфейса могут повлиять на стабильность системы.

---

## Роль оценки риска

Оценка риска используется как дополнительный сигнал для QA-инженера и дополняет AI-анализ изменений интерфейса.

Таким образом система сочетает:

- технический diff интерфейса  
- AI-интерпретацию изменений  
- количественную оценку риска релиза

````python
# ==========================================================
# Release Risk Score
# ==========================================================

added_count = len(added_elements)
removed_count = len(removed_elements)
changed_count = len(changed_elements)

# Простая модель оценки риска
risk_score = (
    added_count * 1 +
    removed_count * 2 +
    changed_count * 3
)

if risk_score < 5:
    risk_level = "LOW"
elif risk_score < 15:
    risk_level = "MEDIUM"
else:
    risk_level = "HIGH"

release_risk = {
    "old_run": run_old,
    "new_run": run_new,
    "added_elements": added_count,
    "removed_elements": removed_count,
    "changed_elements": changed_count,
    "risk_score": risk_score,
    "risk_level": risk_level
}

print("Release Risk Analysis:\n")
print(release_risk)
````

````python
risk_path = os.path.join(
    artifacts_dir,
    "release_risk_analysis.json"
)

with open(risk_path, "w", encoding="utf-8") as f:
    json.dump(release_risk, f, ensure_ascii=False, indent=2)

print("Release risk сохранён:")
print(risk_path)
````

На основе diff интерфейса между версиями система вычисляет простую метрику **Release Risk Score**, которая помогает быстро оценить потенциальный риск изменений и приоритизировать тестирование.
